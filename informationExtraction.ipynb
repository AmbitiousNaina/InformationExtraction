```python
from ast import Attribute
# beautiful soup helps to extract specific data from web page,pip install beautifulsoup4
from bs4 import BeautifulSoup
import requests
# soup.find(Tag)
# soup.find_all(tag)
# soup.get(Attribute)
# soup.text()
# soup.select(css_selector)

# create the obj with the html parser content
response = requests.get("https://example.com")
print(response)
soup = BeautifulSoup(response.text, 'html.parser')
title = soup.title.string
print(title)
```

```python
# get all the links
print("All the links on the page")
for links in soup.find_all('a'):
    href = links.get('href')
    print(href)
```

```python
response = requests.get("https://en.wikipedia.org/wiki/Artificial_intelligence")
soup = BeautifulSoup(response.text, 'html.parser')
title = soup.title.string
print(title)
```

```python
text = soup.find_all('p')
text_content = " ".join([p.text for p in text])
print(text_content[:500])
```

```python
# Named entity recognition
# identifying & categorizing the text - classifying the entities
import spacy
nlp = spacy.load("en_core_web_sm")
doc = nlp("Apple is looking at buying U.K. startup for $1 billion")

for ent in doc.ents:
    print(ent.text, ent.label_)
```
